```python
#Setup 
import pandas as pd 
```

```python
#File to load 
pizza_df = "/Users/shakhnoza/Downloads/pizza_sales.csv"
pizza_data = pd.read_csv(pizza_df)
pizza_data.head()
```

```python
pizza_data.info()
```

```python
pizza_data['order_date']=pd.to_datetime(pizza_data['order_date'], format = '%d-%m-%Y')
print(pizza_data.dtypes)
```

```python
print(pizza_data)
```

```python
pizza_data['order_time']=pd.to_datetime(pizza_data['order_time'],format='%H:%M:%S').dt.time
print(pizza_data.dtypes)
```

```python
print(pizza_data)
```

```python
pizza_data['pizza_size'] = pizza_data['pizza_size'].replace({'S': 'Regular', 'M': 'Medium', 'L': 'Large', 'XL': 'X-Large', 'XXL': 'XX-Large'})
```
